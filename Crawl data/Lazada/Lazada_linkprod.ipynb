```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from time import sleep
import random
import pandas as pd

# Mở trình duyệt Google Chrome
options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--incognito') # Tab ẩn danh
# options.add_argument("--start-maximized") # Full window
options.add_argument('headless') # Không hiển thị chrome
s = Service('../chromedriver.exe')

driver = webdriver.Chrome(service=s, options=options)
driver.implicitly_wait(5)
```

```python
url = [ "https://www.lazada.vn/dien-thoai-di-dong/apple/?service=official",
        "https://www.lazada.vn/dien-thoai-di-dong/apple--samsung/?service=official"]
def crawl(url):       
    with webdriver.Chrome(service=s, options=options) as driver:
        driver.get(url)
        sleep(random.randint(3,5))
        elements = driver.find_elements(By.CSS_SELECTOR,".RfADt [href]") #get link of all items in a page
        sleep(random.randint(3,5))
        title = [elem.text for elem in elements]
        links = [elem.get_attribute('href') for elem in elements]
        dict = {'title':title,'link_product':links}
        df_laz1 = pd.DataFrame(dict)
        return df_laz1

df_laz1 = crawl(url[0])
df_laz2 = crawl(url[1])
df_laz1.to_csv('link_laz1_product.csv', encoding='utf-8')
df_laz2.to_csv('link_laz2_product.csv', encoding='utf-8')
```
